```python
from sklearn import preprocessing
```

```python
g_la=preprocessing.LabelEncoder()
n_la=preprocessing.LabelEncoder()
pob_la=preprocessing.LabelEncoder()
sid_la=preprocessing.LabelEncoder()
gid_la=preprocessing.LabelEncoder()
seid_la=preprocessing.LabelEncoder()
t_la=preprocessing.LabelEncoder()
s_la=preprocessing.LabelEncoder()
r_la=preprocessing.LabelEncoder()
pas_la=preprocessing.LabelEncoder()
pss_la=preprocessing.LabelEncoder()
sad_la=preprocessing.LabelEncoder()
c_la=preprocessing.LabelEncoder()
```

```python
df['gender'] = g_la.fit_transform(df['gender'])
df['NationalITy'] = n_la.fit_transform(df['NationalITy'])
df['PlaceofBirth'] = pob_la.fit_transform(df['PlaceofBirth'])
df['StageID'] = sid_la.fit_transform(df['StageID'])
df['SectionID'] = seid_la.fit_transform(df['SectionID'])
df['GradeID'] = gid_la.fit_transform(df['GradeID'])
df['Topic'] = t_la.fit_transform(df['Topic'])
df['Semester'] = s_la.fit_transform(df['Semester'])
df['Relation'] = r_la.fit_transform(df['Relation'])
df['ParentAnsweringSurvey'] = pas_la.fit_transform(df['ParentAnsweringSurvey'])
df['ParentschoolSatisfaction'] = pss_la.fit_transform(df['ParentschoolSatisfaction'])
df['StudentAbsenceDays'] = sad_la.fit_transform(df['StudentAbsenceDays'])
df['Class'] = c_la.fit_transform(df['Class'])
```

```python
x = df.iloc[:,0:16]
y=df.iloc[:,16]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30 , random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_log = LogisticRegression(max_iter=50)
model_log.fit(x_train, y_train)
```

```python
n_iter = model_log.n_iter_
print(f"Number of iterations used: {n_iter[0]}")
```

```python
y_pred = model_log.predict(x_test)
y_pred_proba = model_log.predict_proba(x_test)[:, 1] 
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
