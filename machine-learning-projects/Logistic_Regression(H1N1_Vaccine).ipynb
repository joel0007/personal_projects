```python
from sklearn import preprocessing
```

```python
ab_la=preprocessing.LabelEncoder()
q_la=preprocessing.LabelEncoder()
r_la=preprocessing.LabelEncoder()
g_la=preprocessing.LabelEncoder()
il_la=preprocessing.LabelEncoder()
ms_la=preprocessing.LabelEncoder()
hs_la=preprocessing.LabelEncoder()
e_la=preprocessing.LabelEncoder()
cmsa_la=preprocessing.LabelEncoder()
```

```python
df['age_bracket'] = ab_la.fit_transform(df['age_bracket'])
df['qualification'] = q_la.fit_transform(df['qualification'])
df['race'] = r_la.fit_transform(df['race'])
df['sex'] = g_la.fit_transform(df['sex'])
df['income_level'] = il_la.fit_transform(df['income_level'])
df['marital_status'] = ms_la.fit_transform(df['marital_status'])
df['housing_status'] = hs_la.fit_transform(df['housing_status'])
df['employment'] = e_la.fit_transform(df['employment'])
df['census_msa'] = cmsa_la.fit_transform(df['census_msa'])
```

```python
x = df3.iloc[:,0:18]
y=df3.iloc[:,18]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30 , random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_log = LogisticRegression(max_iter=100)
model_log.fit(x_train, y_train)
```

```python
n_iter = model_log.n_iter_
print(f"Number of iterations used: {n_iter[0]}")
```

```python
y_pred = model_log.predict(x_test)
y_pred_proba = model_log.predict_proba(x_test)[:, 1] 
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure()
plt.plot(fpr, tpr, label="ROC curve")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.2f}")
```
