```python
df['Time'] = df['Time'].astype(int)
```

```python
def categorize_time(hour):
    if 0 <= hour < 5:
        return 'Late Night'
    elif 5 <= hour < 8:
        return 'Early Morning'
    elif 8 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 16:
        return 'Afternoon'
    elif 16 <= hour < 18:
        return 'Late Afternoon'
    elif 18 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['Char_Time'] = df['Time'].apply(categorize_time)
```

```python
def plot_top_10_authors(ax, df, brand_name):
    top_10_authors = pd.pivot_table(index='author', values='price', aggfunc='sum', data=df).sort_values(by='price', ascending=False).head(10)
    ax.barh(top_10_authors.index, top_10_authors.values.flatten(), color='skyblue')
    ax.set_title(f"Top 10 Authors for {brand_name}")
    ax.set_xlabel("Total Amount Spent")
    ax.set_ylabel("Authors")
    ax.invert_yaxis()
    ax.tick_params(axis='x', rotation=45)

fig, axs = plt.subplots(4, 3, figsize=(16, 14))
plt.subplots_adjust(hspace=0.5, wspace=0.4)
plot_top_10_authors(axs[0, 0], df_olay_data, 'Olay')
plot_top_10_authors(axs[0, 1], df_nykaa_naturals_data, 'Nykaa Naturals')
plot_top_10_authors(axs[0, 2], df_nykaa_cosmetics_data, 'Nykaa Cosmetics')
plot_top_10_authors(axs[1, 0], df_nyx_data, 'NYX Professional Makeup')
plot_top_10_authors(axs[1, 1], df_maybelline_data, 'Maybelline New York')
plot_top_10_authors(axs[1, 2], df_lakme_data, 'Lakme')
plot_top_10_authors(axs[2, 0], df_loreal_data, "L'Oreal Paris")
plot_top_10_authors(axs[2, 1], df_kay_beauty_data, 'Kay Beauty')
plot_top_10_authors(axs[2, 2], df_herbal_data, 'Herbal Essences')
plot_top_10_authors(axs[3, 0], df_colorbar_data, 'Colorbar')
plot_top_10_authors(axs[3, 1], df_nivea_data, 'Nivea')
axs[3, 2].axis('off')
plt.suptitle('Top 10 Authors by Brand', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()
```

```python
def plot_rating_histogram(df, brand_name):
    plt.figure(figsize=(10, 8))
    plt.subplot(1, 2, 1)
    plt.hist(df['product_rating'], bins=10, color='blue', edgecolor='black', alpha=0.7)
    plt.title(f'{brand_name} - Product Ratings Distribution')
    plt.xlabel('Product Rating')
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(df['review_rating'], bins=10, color='green', edgecolor='black', alpha=0.7)
    plt.title(f'{brand_name} - Review Ratings Distribution')
    plt.xlabel('Review Rating')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

plot_rating_histogram(df_olay_data, 'Olay')
plot_rating_histogram(df_nykaa_naturals_data, 'Nykaa Naturals')
plot_rating_histogram(df_nykaa_cosmetics_data, 'Nykaa Cosmetics')
plot_rating_histogram(df_nyx_data, 'NYX Professional Makeup')
plot_rating_histogram(df_maybelline_data, 'Maybelline New York')
plot_rating_histogram(df_lakme_data, 'Lakme')
plot_rating_histogram(df_loreal_data, 'LOreal Paris')
plot_rating_histogram(df_kay_beauty_data, 'Kay Beauty')
plot_rating_histogram(df_herbal_data, 'Herbal Essences')
plot_rating_histogram(df_colorbar_data, 'Colorbar')
plot_rating_histogram(df_nivea_data, 'Nivea')
```
